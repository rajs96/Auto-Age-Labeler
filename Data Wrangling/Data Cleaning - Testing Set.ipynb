```python
# import required modules 
import pandas as pd
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import sys,json, os 
import numpy as np
from pyaudio_featurize import pyaudio_featurize
```

```python
predictor_df = pd.read_csv('cv-valid-test.csv')
predictor_df['filename'] = predictor_df['filename'].apply(lambda s: s[14:])
predictor_df = predictor_df[['filename','age']]
predictor_df.dropna(inplace=True)
predictor_df.head()
```

```python
lst = []
extract_labels = pyaudio_featurize('cv-valid-test/sample-000001.mp3')
labels = extract_labels[1]
for sample in list(predictor_df['filename']):
    features = pyaudio_featurize('cv-valid-test/%s'%(sample))
    row = list(features[0])
    row.insert(0,sample)
    lst.append(row)
    
labels.insert(0,'filename')    
df = pd.DataFrame(lst,columns=labels)
new_df = df.merge(predictor_df,on='filename')
new_df.head(10)
```

```python
new_df.to_csv('audio_testing_data_cleaned.csv') # export cleaned data as csv file
```
